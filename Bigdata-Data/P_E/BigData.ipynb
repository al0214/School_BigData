```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("./Data/Data.csv", encoding='cp949') # 데이터 프레임으로 데이터 가져오기

data = df[['시점', '범죄발생시간별(1)', '계']] # column으로 데이터 짜르기

for a in range(1, len(df["계"]), 10): # 데이터 전처리
    data = data.drop([a])
for b in range(10, len(df["계"]), 10):
    data = data.drop([b])
    
df2011 = data.loc[df['시점'] == "2011"]
df2012 = data.loc[df['시점'] == "2012"]
df2013 = data.loc[df['시점'] == "2013"]
df2014 = data.loc[df['시점'] == "2014"]
df2015 = data.loc[df['시점'] == "2015"]
df2016 = data.loc[df['시점'] == "2016"]
df2017 = data.loc[df['시점'] == "2017"]
df2018 = data.loc[df['시점'] == "2018"]
df2019 = data.loc[df['시점'] == "2019"]
df2020 = data.loc[df['시점'] == "2020"]
df2021 = data.loc[df['시점'] == "2021"]

def divide(a):                              # 데이터 크기로 인하여 크기를 줄임
    a['계'] = a['계'].astype('int') / 10000

bigdata = [df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019, df2020, df2021]

for i in range(0, len(bigdata), 1):
    divide(bigdata[i])

df2020
```

```python
plt.style.use('ggplot')
plt.title("Crime rate by hour")
plt.rc('font', family="AppleGothic")
plt.rcParams['axes.unicode_minus'] = False
plt.rc('legend', fontsize="x-small")
plt.plot(df2011["범죄발생시간별(1)"], df2011["계"], color='red', label = '2011')
plt.plot(df2012["범죄발생시간별(1)"], df2012["계"], color='orange', label = '2012')
plt.plot(df2013["범죄발생시간별(1)"], df2013["계"], label = '2013')
plt.plot(df2014["범죄발생시간별(1)"], df2014["계"], color='green', label = '2014')
plt.plot(df2015["범죄발생시간별(1)"], df2015["계"], color='blue', label = '2015')
plt.plot(df2016["범죄발생시간별(1)"], df2016["계"], color='purple', label = '2016')
plt.plot(df2017["범죄발생시간별(1)"], df2017["계"], color='pink', label = "2017")
plt.plot(df2018["범죄발생시간별(1)"], df2018["계"], color='skyblue', label = "2018")
plt.plot(df2019["범죄발생시간별(1)"], df2019["계"], color='hotpink', label = "2019")
plt.plot(df2020["범죄발생시간별(1)"], df2020["계"], color='aqua', label="2020")
plt.plot(df2021["범죄발생시간별(1)"], df2021["계"], color='tan', label="2021")
plt.xticks(rotation=45)
plt.yticks((np.arange(5, 35, 5)),("5만명", "10만명", "15만명", "20만명", "25만명", "30만명"))
plt.grid(True)
plt.legend()
plt.show()
```
