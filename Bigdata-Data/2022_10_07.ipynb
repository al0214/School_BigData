<h3>In 연산자</h3>
'울산광역시 중구 성안동(3111065000)'처럼 정확하게 지역명을 입력하는 것은 번거로룸.
파이썬의 in 연산자를 이용하여 효율적으로 코드 작성<br>
'A in B'는 A가 B안에 존재하면 '참', 그렇지 않으면 '거짓'
즉 in 연산자는 data의 row[0] 값 중 똑같은 값이 있는지 찾아서,
그 결과를 True 또는 Flase로 반환

```python
print('신도림' in '서울특별시 구로구 신도림동(1153051000)')
print('1153' in '서울특별시 구로구 신도림동(1153051000)')
print('()' in '서울특별시 구로구 신도림동(1153051000)')
```

<h2>데이터 전처리</h2>

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

result = []

f = open('ingu.csv', encoding="cp949")
df = csv.reader(f)

a = input("원하는 위치를 입력해 주세요")

for row in df:
    if a in row[0]:
        for i in row[3:]:
            result.append(int(i))
print(result)

plt.plot(result)
plt.show()
```
