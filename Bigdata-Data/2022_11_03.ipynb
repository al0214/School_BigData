```python
import csv
import pandas as pd

f = open('./gender.csv', encoding="cp949")
df_data = csv.reader(f)
next(df_data)

a = input("원하는 위치를 입력하여주세요")
m = []
f = []

for row in df_data:
    if a in row[0]:
        for i in range(0, 101):
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))
        break
f.reverse()
m.reverse()

print(f)
print(m)

import matplotlib.pyplot as plt
plt.barh(range(101), f)
plt.barh(range(101), m)
plt.show()
```

```python
size = [2441, 2312, 1031, 1233]
name = ['A형', 'B형', 'AB형', 'O형']
# 원 그래프라고도 불리는 pie() 함수는 전체 데이터 중 특정 데이터의 비율을 보기 쉽게 표현함
plt.rc('font', family = 'appleGothic')
plt.axis('equal') # x축과 y축의 비율을 실제 똑같은 비율로 나오게 하는 메소드
plt.pie(size, labels=name, autopct='%.1f%%')
plt.legend()
plt.show()

# autopct 속성은 auto percent를 의미
# 어떤 값으로 표시할지 작성하면 각 항목의 비율을 자동으로 계산해서 표시
# 소수점 아래 둘째 자리에서 반올림한 값을 표시하고자 autopct 값을 %.1f%%로 지정
# 특정 비율의 돌출 효과 explode ()는 돌출되지 않음
# 옵션에 대한 자세한 내용은 matplotlib.org 참고
```

```python
import csv
import pandas as pd

f = open('./BloodType.csv')
data = csv.reader(f)
next(data)

A = 0
B = 0
AB = 0
O = 0

for row in data:
    if(row[2]!='혈액형'):
        if(row[2]=='A'):
            A += 1
        if(row[2]=='B'):
            B += 1
        if(row[2]=='AB'):
            AB += 1
        if(row[2]=='O'):
            O += 1
plt.rc('font', family = 'appleGothic')
plt.axis('equal') # x축과 y축의 비율을 실제 똑같은 비율로 나오게 하는 메소드
plt.pie([A,B,AB,O], labels=name, autopct='%.1f%%')
plt.legend()
plt.show()
   
       
    
       
            
```

```python
import csv
import pandas as pd

f = open('./BloodType.csv')
data = csv.reader(f)
next(data)

bloodtype = []

for row in data:
    if(row[2]!='혈액형'):
        bloodtype.append(row[2])

plt.rc('font', family = 'appleGothic')
plt.axis('equal') # x축과 y축의 비율을 실제 똑같은 비율로 나오게 하는 메소드
plt.pie(
    [
        bloodtype.count('A'),
        bloodtype.count('B'),
        bloodtype.count('AB'),
        bloodtype.count('O')
     ], 
    labels=name, 
    autopct='%.1f%%'
    )
plt.legend()
plt.show()
```

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

f = open('./gender.csv', encoding="cp949")
df_data = csv.reader(f)
next(df_data)

a = input("원하는 위치를 입력하여주세요")
size = []

for row in df_data:
    if a in row[0]:
        m = 0
        f = 0
        for i in range(0, 101):
            m += int(row[i+3])
            f += int(row[i+106])
size.append(m)
size.append(f)
print(size)

name = ['남자', '여자']

plt.title(f'{a} 남녀 비율')
plt.rc('font', family = 'appleGothic')
plt.axis('equal')
plt.pie(size, labels=name,
    autopct='%.1f%%',
    startangle = 90
    )
plt.legend()
plt.show()
```
