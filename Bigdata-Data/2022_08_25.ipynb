```python
# numpy         .
x = np.random.randint(-100, 100, 100) # -100 100   100 (x)
y = np.random.randint(-200, 200, 100) # -200 200   100 (y)
plt.scatter(x,y)
```

```python
#    x, y     
# size    .
size = np.random.randint(-200,200,100)
plt.scatter(x,y,s=size)
```

```python
#   c    alpha     .
size = np.random.randint(-200,200,100)
size = np.random.randint(0,100,100)
plt.scatter(x,y,s=size, c=x ,alpha=0.5)
plt.colorbar()
```

```python
x = np.random.randint(-100, 100, 1000) # 1000   
y = np.random.randint(-100, 100, 1000) # 1000   
size = np.random.randint(100)

mask1 = abs(x) > 50 # x     5   
mask2 = abs(y) > 50 # x     5   
x = x[mask1+mask2] # mask1 mask2     
y = y[mask1+mask2] # mask1 mask2     

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
