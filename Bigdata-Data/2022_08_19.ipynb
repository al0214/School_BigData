```python
# 데이터의 시각화-matplotlib
# matplotlib는 파이썬의 대표적인 과학 계산용 그래프 라이브러리
# 탐색적 데이터 분석과 데이터 분석과 데이터 결과의 시각화를 위해 자주 사용됨
# 그래프를 그릴때는 pyplot 부속 라이브러리를 사용하고 별칭은 관례적으로 plt 사용

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data1.csv', encoding='cp949')

plt.plot([10, 20, 30, 40])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.title('plotting')
plt.plot([1, 2, 3,4], [12, 42, 25, 15])
plt.show()
```

```python
# 범례 표시 - 그래프가 의미하는 바를 구별하기 위한 방법
# 볌례는 보통 두개 이상의 데이터를 표시할ㄷ 때 사용합니다.
# plot() 함수에 label이라는 속성이 레이블 값으로 원하는 문자열을 넣어주고
# 그래프를 그리기 전에 legend() 함수를 실행 시키면 레이블 값이 범례로 나타남

plt.title("legend")
plt.plot([10, 20, 30, 40], 'skyblue',linestyle='--', label="asc")
plt.plot([40, 30, 20, 10], 'pink', label="dasc")
plt.legend()
plt.show()
```

```python
# 마커 모양 바꾸기
# plot() 함수에 marker 속성을 설정하면 선 형태가 아닌 점 형태로 그래프를 그릴 수 있음.
# 이때 색상과 마커 모양을 한번에 설정 할 수도 있음
# 마커 모양으로는 점, 선, 동그라미, 삼각형, 별모양 등 여러가지가 있음

plt.title('marker')
plt.plot([10, 20, 30, 40], 'r.', label='circle')
plt.plot([40, 30, 20, 10], 'g^', label='triangle up')
plt.legend()
plt.show()
```
